```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# 1. Load the dataset from GitHub URL
data_url = "https://raw.githubusercontent.com/kdemertzis/EKPA/main/Data/pcap_data.csv"
data = pd.read_csv(data_url)
```

```python
# 2. Data Preprocessing
# Check for missing values
data = data.dropna()
```

```python
# Separate features and labels
# Check if 'Label' column exists before dropping  # Changed from 'label' to 'Label'
if 'Label' in data.columns:
    X = data.drop(columns=['Label'])
    y = data['Label']
else:
    # Handle the case where 'Label' column is not found
    # For example, print an error message or try to find the correct column name
    print("Error: 'Label' column not found in the DataFrame.")
    # You can print the available columns for debugging:
    print("Available columns:", data.columns)
    # You might need to adjust the column name if it's different
    # For example, if the target column is named 'target', change the code accordingly:
    # Assuming the target column is named 'Label'
    # **The column name in the dataset is actually "Label", not "label"**
    # X = data.drop(columns=['Label'])  # Replace 'Label' with the actual target column name if different
    # y = data['Label']  # Replace 'Label' with the actual target column name if different
    # Assuming the label column is actually named 'Avg_syn_flag' or similar:
    # Check data types of columns to identify potential label column
    print(data.dtypes)
    # Choose a suitable categorical column for the label (e.g., 'Protocol')
    # The original code assumed the label column was 'Protocol'.
    # Replace 'Protocol' with the actual label column name from data.columns.
    # Based on the provided context, 'Avg_syn_flag' seems like a potential label column
    label_column = 'Avg_syn_flag'  # Replace with actual label column name from data.columns
    X = data.drop(columns=[label_column])
    y = data[label_column]
```

```python
# Encode categorical variables if necessary
X = pd.get_dummies(X, drop_first=True)
```

```python
# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 4. Train a Random Forest Classifier
# Assuming 'Avg_syn_flag' contains continuous values, you might need regression instead of classification
from sklearn.ensemble import RandomForestRegressor  # Import Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)  # Use Regressor
model.fit(X_train, y_train)
```

```python
# 5. Evaluate the Model
# Import necessary functions
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)

# Calculate Regression Metrics
mse = mean_squared_error(y_test, y_pred)
# Calculate RMSE by taking the square root of MSE
rmse = mse**0.5  # Or you can use np.sqrt(mse) if you import numpy
r2 = r2_score(y_test, y_pred)

# Print regression metrics
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
# 6. Save the model (optional)
import joblib
joblib.dump(model, 'intrusion_detection_model.pkl')
```
